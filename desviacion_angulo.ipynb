```
import matplotlib.pyplot as plt
import numpy as np
```

```
a = np.array([0.1]*50)
b = np.sin(np.linspace(0,4,50)) + np.random.uniform(-0.1,0.1,50)
c = np.cos(np.linspace(2,6,50)) + np.random.uniform(-0.1,0.1,50)
d = [i - np.random.uniform(0,0.8) for i in np.linspace(0,1,50)]
```

```
fig,ax = plt.subplots(1,1,figsize=(15,7),dpi=120)
ax.plot(np.linspace(0,50,50),a, marker='o', linestyle='dashed', label='Inversión a')
ax.plot(np.linspace(0,50,50),a, marker='o', linestyle='dashed', label='Inversión b')
ax.plot(np.linspace(0,50,50),a, marker='o', linestyle='dashed', label='Inversión c')
ax.plot(np.linspace(0,50,50),a, marker='o', linestyle='dashed', label='Inversión d')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Fracción de retorno')
ax.legend()
plt.show()
```
